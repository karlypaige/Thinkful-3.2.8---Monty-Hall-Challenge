You are on a game show and given the choice of whatever is behind three doors. Behind one door is a fantastic prize (some examples use a car, others use cash) while behind the other two doors is a dud (some examples say a goat, others say it's just empty). You pick a door. Then the host opens one of the other two doors to reveal a dud. But here's the wrinkle: the host now gives you the opportunity to switch your door. What should you do?

Intuitively speaking, initially you had a 2 in 3 chance of choosing incorrectly.  When one choice was eliminated after you selected it didn't change the initial probability. So it is still more likely that you initially chose wrong and switching after the wrong door is eliminated is the best choice.

Attempting to put this into Bayesian terms (before peeking at the answer):

P(wrong|elimination)	= 1 * (0.66/1) = 0.66
P(elimination|wrong)	= 1	because one was certainly going to be eliminated
P(wrong)	= 0.66	
P(elimination)	= 1	because one was certainly going to be eliminated
		
P(right|elimination)	= 1 * (0.33/1) = 0.33	
P(elimination|right)	= 1	because one was certainly going to be eliminated
P(right)	= 0.33	
P(elimination)	= 1	because one was certainly going to be eliminated
